```r
# 1.2. Типы данных

# Из результатов функций spec, glimpse и skim_tee мы видим, что в таблице dur_word_freq используются данные двух классов -- строковые (character) и непрерывные числовые (numeric типа double)
```

```r
# 1.3 Визуализации ggplot: график плотности (density plot)

dur_word_freq |>
ggplot(aes(x = LogDurationA, group = Cond)) +
geom_density()
```

```r
# 1.4 Добавьте в график тему theme_classic() и полупрозрачность (alpha = 0.5, указывается в геоме плотности).

dur_word_freq |>
ggplot(aes(x = LogDurationA, group = Cond, fill=Cond)) +
geom_density(alpha=0.5) +
theme_classic()
```

```r
# 1.5 Постройте график плотности распределения длины всего слова (переменная LogDurationW) для всех данных (без деления по условиям).

dur_word_freq |>
ggplot(aes(x = LogDurationW)) +
geom_density() +
theme_classic()
```

```r
# 1.6 Боксплот (базовый R)

boxplot(LogDurationA ~ Cond, data=dur_word_freq)
```

```r
# 1.7 Сравните положение медианных значений LogDurationA, а также значений 1-го и 3-го квартиля в условиях fast и slow, запишите кратко ваши выводы ниже (2-3 предложения).

# Медианная длина звука в условии slow меньше, чем в условии fast. Интерквартильный размах в условии slow меньше, чем в условии fast, что говорит о меньшем разбросе даннныхэ
```

```r
# 1.8 Боксплот в ggplot2

library(RColorBrewer)

dur_word_freq |>
ggplot(aes(y = LogDurationW, x=Cond, fill=Cond)) +
geom_boxplot() +
theme_classic() +
scale_fill_brewer() +
xlab("Condition") + ylab("Log-transformed word duration") + ggtitle("Word duration in slow and fast conditions")
```

```r
# 1.9 Скрипичный плот в ggplot2

p <- dur_word_freq |>
ggplot(aes(y = LogDurationA, x=Cond, fill=Cond)) +
geom_violin(alpha=0.3) +
theme_classic() +
xlab("Condition") + ylab("Log-transformed segment duration") + ggtitle("Segment duration in slow and fast conditions")

p
```

```r
# 1.10 Jitter

p + geom_jitter(aes(col=Cond))
```

```r
# 1.11 Скаттерплот в ggplot2

dur_word_freq |>
ggplot(aes(x = LogDurationA, y=LogDurationW)) +
geom_point(alpha=0.3) +
theme_classic() +
geom_smooth(method='lm', se = FALSE) +
xlab("Log-transformed word duration") + ylab("Log-transformed segment duration") + ggtitle("Word vs. segment duration")
```
